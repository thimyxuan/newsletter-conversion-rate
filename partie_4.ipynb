## Résultats du modèle

Le modèle de prédictions maximise l'importance du nombre de pages visitées pour déterminer si un utilisateur s'inscrira ou non à la newsletter.

Cela confirme ce que l'on avait observé dans l'analyse des données en partie 1.

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings('ignore', category=UserWarning)

feature_importance = pd.read_csv('./src/feature_importance.csv')
sns.set(rc={'figure.figsize':(6, 2)}, style='white')
sns.barplot(data = feature_importance, orient = 'v', palette = ['limegreen'], width = .9, edgecolor = 'black')
plt.title('Features importance in the Random Forest Model', fontsize=12.5, pad=20)
plt.xlabel('', fontsize=10)
plt.ylabel('Coefficients', fontsize=10)
plt.xticks(fontsize=9, fontweight='bold', rotation=45, ha='right')
plt.xticks(feature_importance.columns, fontsize=9, fontweight='bold', rotation=45, ha="right")
plt.legend('', frameon=False)
plt.show()
```

## Conclusion

- Le taux de conversion est faible. Moins de 2 personnes sur 100 qui visitent le site web s'abonnent à la newsletter.  

- Ce taux peut s'en doute s'améliorer.

- Les personnes qui s'abonnent à la newsletter présentent un intérêt pour le site web.  

- Elles visitent en moyenne 3 fois plus de pages que le groupe total.  

--> <ins>Hypothèse</ins> : Plus un utilisateur visite de pages, plus il a de chance de s'abonner à la newsletter.
